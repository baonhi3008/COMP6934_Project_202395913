# Introduction

**The Client**

The International Energy Transition Advisory Group (IETAG) is a global think tank specializing in energy policy, sustainability, and social equity. Their core challenge lies in balancing economic development with environmental responsibility and equitable energy access—issues that are inherently interconnected and require deep, multidimensional analysis.

**The Dataset**

To support this mission, IETAG has curated a set of approximately eight global datasets spanning 1990–2023. These include indicators on GDP, urbanization, energy consumption, CO₂ emissions, electricity and clean fuel access, and the renewable energy share in electricity generation. Together, these datasets offer a comprehensive view of how economic and demographic development intersects with environmental outcomes and social equity in energy systems. The data is drawn from trusted sources such as the World Bank, OWID, and IEA.

By leveraging interactive data visualization techniques, this project aims to go beyond descriptive statistics to uncover spatial, temporal, and relational patterns that may be difficult to observe in raw or tabular form. This approach enables stakeholders to engage with the data more intuitively and investigate both global trends and regional disparities.

**Overall Goal**

Project statement **"Exploring the Interplay of Economic Growth, Energy Justice, and Sustainability in the Global Energy Transition"**  
This project aims to develop an in depth understanding of the relationship of economic growth, energy justice, and sustainability in the global energy transition to inform decision making for policymakers, energy companies, and social equity advocates. 

**Objectives & Analytical Focus Areas for below visual design**
1. Analyze the impact of economic growth and urbanization on energy consumption and emissions:

    - Examine how rising urbanization and GDP influence regional energy demand and CO₂ emissions.
    - Explore whether shifts in the energy mix signal meaningful progress in emissions mitigation.

    - Assess the relationship and trends between economic growth and energy consumption has changed from 1993 to 2023.

3.	Identify energy access disparities and evaluate the role of renewables in addressing them:

    - Explore whether GDP per capita growth leads to real gains in electricity and clean fuel access.

```python

regional_mix_df = df_main.groupby(['subregion', 'year']).agg({
    'coal_share_energy': 'mean',
    'gas_share_energy': 'mean',
    'oil_share_energy': 'mean',
    'renewables_share_energy': 'mean'
}).reset_index()

regions = regional_mix_df['subregion'].dropna().unique()

num_regions = len(regions)
cols = 4  
rows = (num_regions + cols - 1) // cols  

fig, axes = plt.subplots(rows, cols, figsize=(20, rows * 5), sharex=True, sharey=True)

axes = axes.flatten()

for i, region in enumerate(regions):
    region_data = regional_mix_df[regional_mix_df['subregion'] == region]
    ax = axes[i]
    ax.stackplot(region_data['year'],
                    region_data['coal_share_energy'],
                    region_data['gas_share_energy'],
                    region_data['oil_share_energy'],
                    region_data['renewables_share_energy'],
                    labels=['Coal', 'Gas', 'Oil', 'Renewables'],
                    colors=sns.color_palette("Set2"))
    ax.set_title(region)
    ax.set_xlabel("Year")
    ax.set_ylabel("Average Share (%)")

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

fig.legend(['Coal', 'Gas', 'Oil', 'Renewables'], loc='upper center', ncol=4, fontsize=12)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.suptitle("Evolution of Energy Source Mix Over Time by Subregion", fontsize=16)
plt.show()
```

### **Observations**  

**Data Completeness:**  

- <mark>  	The data is real world data, and as such, it naturally contains numerous missing values. These missing values are especially evident in many variables such as primary_energy_consumption, CO₂, electricity_demand_per_capita,...</mark>
- Especially data is often missing in countries that is classified with low incomes such as regions like Africa and Asia often have significant gaps—many years of missing entries simply due to the challenges in data collection. I decide to NOT impute these values preserves the contextual integrity of the dataset that can ensure the subsequent analyses reflect the true state of the data.
  

**Energy Transition Perspective:**  

- The dataset is inherently valuable for assessing the progress in transitioning from fossil fuels to renewable energy sources. Although missing values exist, particularly in regions with historical data collection challenges, these very gaps provide meaningful context. For instance, the absence of data from certain low‑income countries signals the real challenges faced in documenting energy consumption and emissions accurately.
- <mark>When analyzing trends in fossil fuel share versus renewable share, it is critical to consider that some countries may have extensive gaps. </mark>. This approach prevents the imputation from obscuring underlying issues and ensures that the assessment of the energy transition remains grounded in the actual availability and quality of data.
- Keeping the missing values allows for an honest evaluation of where improvements occur and where data collection needs to be strengthened, particularly in emerging economies in Africa and Asia.

# How Analysis

### Each dashboard is custom designed around a specific analytical question , integrating multiple views and interactive features that I implemented manually. While some components were inspired by open-source examples, I do a lot of customization and extended them with derived metrics, region specific filtering, and multilevel interactions that reflect my own task and data abstraction design. These visualizations are not directly replicable with templates and were developed iteratively with attention to both narrative flow and exploratory usability

## Justification and motivation in general: 

During planning and conceptualizing process for the visual design for this project, I quickly I quickly realized that using just a single chart, no matter how well-designed, wouldn’t be enough. For example in visual report of TrackingSDG7 and OWID websites, some only have map or chart as a single view which some how requires extra step and ability to memorize things, which reduce the ability of observation and immediate recognition of insights. Therefore with all the questions I’m addressing, I choose the multiview with customized filtering logic, layout and add in many interaction + visual components within the common visual design such as map. Additionally, the data required for my analytical questions are multidimensional and involve complex relationships between variables like GDP, energy consumption, emissions, electricity access, and urbanization. These aren’t simple one variable comparisons; they require users to explore patterns, track changes over time, compare across regions, and identify outliers or correlations.
From what I’ve seen in many existing energy related visualizations, it’s common to use dashboards with multiple coordinated views and interactivity. These designs are often necessary, not just a choice of styling. They help users make sense of complex data by allowing them to look at it from different angles simultaneously. That influenced my decision here too. By integrating multiple views such as choropleth maps, time series, scatter plots, and bubble overlays, I can give users a way to navigate both high level patterns and specific details. The coordinated design makes it easier to compare variables, spot anomalies, and build mental connections between related metrics. Moreover, interaction plays a big role too: filtering, toggling, or drilling down helps reveal insights that wouldn’t surface in a static or single view format. In the end, I wanted to create a dashboard that feels exploratory and flexible, not just presentational. This approach aligns well with Tamara Munzner’s nested model of visualization and with Shneiderman’s mantra: “overview first, zoom and filter, then details on demand.” In this way, the dashboard serves as a cognitive tool that supports reasoning, hypothesis generation, and decision making across a range of energy related questions.


- **For question 1:**

When I think about how to explore the evolution of energy consumption and CO₂ emissions alongside urbanization and GDP growth, I realize that a single chart just wouldn’t be enough. These variables interact across time and space. Additionally, the analysis considers whether shifts in the energy mix signal efforts toward mitigation is worth exploring. Those are good justifications to design a dashboard with multiple coordinated views: one part shows a spatial overview of GDP or urbanization growth, while bubble overlays and time series reveal patterns in energy use and emissions. This design allows user to compare regions, monitor temporal trends, and examine where renewable adoption may be making an impact, such integrative exploration would not be possible through a single design element. 

- **For question 2:**

For this question, I’m trying to understand how the relationship between economic growth and energy consumption has changed over time—and not just globally, but across different regions and income groups. I also want to see which countries are outliers. To do that, I need more than one view. Using a multipleview + spatial encoding and combination of maps, scatter plots, and time series lets me look at long-term trends, regional shifts, and country level anomalies. If I just used a single chart, I’d either lose the time aspect, the spatial aspect, or the ability to compare groups. Multiple views help me (also myb cliebt) to keep all of that context visible so users can dig into both patterns and exceptions. 

- **For question 3:**

I’m interested in figuring out where electricity access is still lacking, especially in rural areas—and whether a higher share of renewables actually helps improve the situation. This involves looking at geographic differences, comparing urban and rural access, and bringing in renewables data too. Since that’s already three variables across space and time, a single chart would oversimplify things. By using a dashboard, I can look at maps for spatial context, compare urban vs. rural rates side-by-side, and use scatter plots to see how renewables relate to access. It gives me a more complete picture and makes it easier to explore the patterns.

-  **For question 4:**

In this question, I’m trying to connect economic growth with improvements in electricity and clean fuel access. It’s not just about showing that GDP is rising, it’s about whether that’s actually translating into better access rate and improve basic living condition. Instead, I use coordinated views to explore how electricity access and clean cooking fuel access move with GDP per capita. **a.Success Critique**

In overall, this visualization gives strong visual differentiation that satisfy multiple level of analysis from low to high level with many visual design from complicated like choropleth+bubble overlay to simple one like line charts. Additionally, it also offers a country specific overview with very detail information which can be indicates as the detail card when user clicked on specific country.

**Pros**
 
- Multi-View Interactivity:

	- The dashboard integrates several interactive components, including a map with choropleth shading and bubble overlays, two scatterplots, two time series line charts, and a series of stacked area charts arranged in facets.
	The interaction spans across all views rather than being restricted to individual elements, ensuring coordinated updates and filtering driven by a central interactive function. This comprehensive integration is an essential and creative feature of the design.

- Smooth Animated Transitions:

	- The use of Plotly transitions enhances the user's experience by smoothly animating chart updates, such as adjustments made through the year slider or dropdown selections. These animations help users to perceive changes intuitively and maintain their cognitive context.


- Detail Card with Linked Interaction:

	- When users select a bubble on the map, a detailed information card dynamically appears, providing specific visualizations for that country, including a pie chart, dual line chart, and a stacked area chart.
	The detail card further demonstrates advanced interactivity through its responsiveness to ongoing inputs. For instance, if users adjust the slider while the card is open, the visualizations within the card update accordingly. This sophisticated coordination enhances the analytical capability and depth of user interaction.

**Cons:** 
- No outlier detection: Outlier countries, for instance low energy but high emissions, are not automatically highlighted                  
- No customized user defined metrics: Users can’t create or view custom ratios like `CO₂ / renewables share` -> less interactive                                 
- Limited legend clarity 
- Design is pretty complex and user must be actively perceive information or it might get them confuse.

From a utility perspective, the dashboard provides considerable value for analysts and policymakers interested in energy policy, environmental impact, with extra context of economic development. The seamless visual transitions also improve perception, making temporal and spatial data shifts intuitively understandable.

**c. Suggested improvements:**

- **Enhanced anomaly detection:** Currently, the visualization lacks explicit highlighting or alerts for countries that deviate significantly from expected patterns, for example country has high GDP but low energy consumption or high emissions. Incorporating anomaly detection, perhaps via color highlights or visual markers, would immediately draw attention to outliers which adds analytical value.

- **User defined metrics and customization:** The current visualization does not permit custom user defined derived metrics. - Clustering/Pattern insight: Lacks advanced grouping, such as clustering countries with similar access/economic profiles.
- No storytelling guide: Users may not know where to start or what to look for on first use.


**b. Usefulness, Utility, and Perception**  

The dashboard strikes a good balance between accessibility and depth, with clear color scales, intuitive overlays, and filtering controls make it approachable for both technical and non-technical audiences. Users can explore both surface level metrics and deeper analytical questions like trends, outliers, and correlations—without feeling overwhelmed. Consistent visual encoding and thoughtfully chosen defaults support pattern recognition and ease of interpretation.



**c. Suggested Improvements**  
 
- **Clarify Legends:** Improve or replace placeholder traces in scatter plots and overlays with more informative, context-rich legends to aid interpretation.  
- **Include Derived Metrics:** Precompute and display key indicators like the urban–rural access gap or yearly improvement rate to help users quickly grasp trends.  
- **Enhance Mobile Responsiveness:** Optimize layouts, especially for faceted charts, to ensure usability across a range of screen sizes.  
- **Add Progress Feedback:** For slower data loads or complex queries, include subtle visual feedback like a progress bar to keep users informed.



# 6. Conclusions

### Success of Visual Designs  
There are a total of four visual designs throughout this project, which include many visual elements that effectively addressed the stated goals. These are driven by data specific questions about global energy across many aspects which ranges from consumption and emissions to accessm, in the context of socio-economic development. Each dashboard thoughtfully combines geographical, temporal, and relational views, such as choropleth maps, animated scatter plots, combined plots, dual plots, faceted view charts, and detailed drill down panels—that comprehensively present a cohesive and interactive analytical experience. Users are empowered to navigate between global overviews and fine grained country level trends, revealing insights about energy demand, generation and environmental impact, inequality in access, the role of renewables, and shifts in energy consumption patterns.  

This multiview coordination, tightly coupled with intuitive interaction design, greatly supports user exploration and interpretation. Nonetheless, a few usability aspects could be further improved. One major limitation is that the complexity of the design may confuse users who are not tech savvy. Additionally, the lack of certain interactions can potentially lead to misinterpretation. There are also many aspects that could be optimized to create more complete visualizations for expert users, such as legend clarity, customizable metrics, additional interaction within encoded elements, and improved visibility of critical patterns. These issues must be addressed in the future to fully support a wider range of users with diverse abilities and technical backgrounds.  

Moreover, the current panel heavy layout is not ideal for smaller screens such as tablets or phones. The dashboard can be redesigned to combine multiple elements into single visuals and reduce the number of split panels. However, this must be carefully considered to develop a simpler yet more meaningful layout. For the time being, the author has done their best to visualize the stated research questions and topics.


### Critique of Methodology  
The design process and results strongly followed Tamara Munzner's methodology, concepts, and frameworks; particularly her "what," "why," and "how" structure—which helped maintain a principled approach to task abstraction, visual encoding, and idiom selection. This ensured each design was tightly aligned with the underlying data types and user actions. The framework was useful in producing rigorous and consistent analysis across the four visual designs (dashboards).  

Although Munzner's framework offers comprehensive and clear guidelines, it leans heavily toward a systematic abstraction of tasks and data, which sometimes lacks emphasis on storytelling and engagement. In addition to Munzner's methodology, several complementary approaches were taken into consideration, such as Shneiderman’s mantra for interaction flow, cognitive load theory, and user-centered design principles—which proved helpful in bridging this gap and informing choices related to layout, filtering, and onboarding.  

Additionally, I was occasionally confused by some of the taxonomy and methodologies introduced by Munzner, so in the classification and analysis there may be some confusions or mistakes.


### Recommended Improvements  
Overall, for future improvements and suggestions, the following key areas can be addressed:

1. **Accessibility** can be enhanced by offering alternative color palettes to support users with vision deficiencies, and by applying high contrast palettes across views.  
2. **More analysis and derived metrics** should be considered—such as renewables effectiveness ratios or emissions indicators versus energy consumption entropy—to improve data interpretability without requiring users to compute values manually.  
3. **Automated insights**, such as highlighting anomalies or labeling significant milestones, would help users discover patterns that might otherwise be overlooked.  
4. **Storytelling elements** such as narrative panels, click and explore interactions, or annotated transitions would provide context and help guide users through the data.  
5. **Responsive layout behavior** across devices, and a redesign to reduce excessive paneling, would broaden accessibility and make the dashboard more usable in real world settings.  
6. **Customization mechanisms** that allow users to define their own metrics would further support personalized exploration and domain specific questions.


In summary, the designs represent a meaningful and well-grounded step toward visualizing energy equity and environmental performance. By continuing to refine these systems through the lens of visual cognition, interaction design, and inclusive storytelling, future versions can offer not only deeper analytical power but also broader accessibility and engagement.
